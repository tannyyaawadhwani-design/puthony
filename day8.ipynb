```python
#basic of visualization matplotlib
import matplotlib.pyplot as plt
import pandas as pd

# Sample DataFrame
data = {"Name": ["Tanya", "Shivani", "Rishi", "Shubham"],
        "Marks": [99, 88, 77, 66]}
df = pd.DataFrame(data)


#Line Plot
plt.plot(df["Name"], df["Marks"])
plt.title("Marks of Students")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()


#Bar Plot
plt.bar(df["Name"], df["Marks"])
plt.title("Marks of Students (Bar Chart)")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()


#Scatter Plot
plt.scatter(df["Name"], df["Marks"])
plt.title("Marks Scatter Plot")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()


#Pie Chart
plt.pie(df["Marks"], labels=df["Name"], autopct="%1.1f%%")
plt.title("Marks Distribution")
plt.show()


#MINI CHALLENGE
#Bar Plot
plt.bar(df["Name"], df["Marks"])
plt.title("STUDENT MARKS COMPARISON")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()


#CUSTOMIZE BAR PLOT
#CHANGE BAR COLOR
plt.bar(df["Name"], df["Marks"], color="pink")
plt.title("STUDENT MARKS COMPARISON")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()


#ROTATE X-AXIS,Y-AXIX LABELS
plt.bar(df["Name"],df["Marks"],color="black")
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.title("STUDENT MARKS COMPARISON")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()


#ADD GRIDLINES
plt.bar(df["Name"], df["Marks"], color="green")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.title("Student Marks Comparison")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()

#ADD LABELS ON BARS
bars = plt.bar(df["Name"], df["Marks"], color="purple")
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2,
             yval + 1, yval, ha="center")

plt.title("Student Marks Comparison")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()

# Create bar plot
bars = plt.bar(df["Name"], df["Marks"], color="teal")

# Add labels automatically
plt.bar_label(bars, label_type="edge", padding=2)

# Titles
plt.title("Student Marks Comparison")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()

#Bigger & Bold Labels

bars = plt.bar(df["Name"], df["Marks"], color="orange")

# Customize font size & weight
plt.bar_label(bars, label_type="edge", padding=3,
              fontsize=12, fontweight="bold")

plt.title("Student Marks Comparison")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()

#Add % Instead of Exact Numbers
bars = plt.bar(df["Name"], df["Marks"], color="green")

# Convert labels to percentage of total marks
total = df["Marks"].sum()
percentages = [f"{(mark/total)*100:.1f}%" for mark in df["Marks"]]

plt.bar_label(bars, labels=percentages, label_type="edge",
              padding=3, fontsize=11)

plt.title("Marks as Percentage of Total")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()

#Custom Format (e.g., "Score: 99")
bars = plt.bar(df["Name"], df["Marks"], color="purple")

labels = [f"Score: {mark}" for mark in df["Marks"]]
plt.bar_label(bars, labels=labels, label_type="edge", padding=3,
              fontsize=10, color="blue")

plt.title("Custom Labels on Bars")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()


#Horizontal Bar Chart
#Sometimes names are long → use barh().

plt.barh(df["Name"], df["Marks"], color="skyblue")
plt.title("Student Marks (Horizontal Bar)")
plt.xlabel("Marks")
plt.ylabel("Name")
plt.show()

#GROUPED BAR CHART
import numpy as np

data = {
    "Name": ["Tanya", "Shivani", "Rishi"],
    "Midterm": [80, 70, 90],
    "Final": [99, 88, 77]
}
df2 = pd.DataFrame(data)

x = np.arange(len(df2["Name"]))  # positions
width = 0.3  # bar width

plt.bar(x - width/2, df2["Midterm"], width,
        label="Midterm", color="orange")
plt.bar(x + width/2, df2["Final"], width,
        label="Final", color="blue")

plt.xticks(x, df2["Name"])
plt.title("Midterm vs Final Marks")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.legend()
plt.show()


#Stacked Bar Chart
#Show contribution inside one bar.

plt.bar(df2["Name"], df2["Midterm"], label="Midterm",
        color="green")
plt.bar(df2["Name"], df2["Final"], bottom=df2["Midterm"],
        label="Final", color="purple")

plt.title("Stacked Bar Chart")
plt.xlabel("Name")
plt.ylabel("Total Marks")
plt.legend()
plt.show()


#Different Colors for Each Bar
colors = ["red", "pink" ,"green", "purple"]
plt.bar(df["Name"], df["Marks"], color=colors)
plt.title("Student Marks (Different Colors)")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()


#Basic Line Plot

import matplotlib.pyplot as plt
import pandas as pd

data = {"Day": [1, 2, 3, 4, 5],
        "Marks": [65, 70, 80, 90, 85]}
df = pd.DataFrame(data)

plt.plot(df["Day"], df["Marks"])
plt.title("Student Marks Over Days")
plt.xlabel("Day")
plt.ylabel("Marks")
plt.show()

#Customize Line (Color, Style, Width)

plt.plot(df["Day"], df["Marks"], color="red",
         linestyle="--", linewidth=1)
plt.title("Customized Line Plot")
plt.xlabel("Day")
plt.ylabel("Marks")
plt.show()

#Add Markers (dots, squares, stars, etc.)

plt.plot(df["Day"], df["Marks"], marker=".", color="blue",
         linestyle="-")
plt.title("Line Plot with Markers")
plt.xlabel("Day")
plt.ylabel("Marks")
plt.show()



#Multiple Lines in One Plot

df2 = pd.DataFrame({
    "Day": [1, 2, 3, 4, 5],
    "Marks_Tanya": [65, 70, 80, 90, 85],
    "Marks_Shivani": [60, 68, 75, 85, 80]
})

plt.plot(df2["Day"], df2["Marks_Tanya"],
         marker="o", label="Tanya", color="pink")
plt.plot(df2["Day"], df2["Marks_Shivani"],
         marker="s", label="Shivani", color="black")

plt.title("Comparison of Marks Over Days")
plt.xlabel("Day")
plt.ylabel("Marks")
plt.legend()
plt.show()


#ADD GRIDLINES
plt.plot(df["Day"], df["Marks"], marker="o", color="green")
plt.grid(True, linestyle="--", alpha=0.6)
plt.title("Line Plot with Gridlines")
plt.xlabel("Day")
plt.ylabel("Marks")
plt.show()



#Basic Scatter Plot

import matplotlib.pyplot as plt
import pandas as pd

#Sample Data
data = {"Study_Hours": [2, 3, 4, 5, 6, 7],
        "Marks": [50, 55, 65, 70, 78, 90]}
df = pd.DataFrame(data)

plt.scatter(df["Study_Hours"], df["Marks"])
plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show


#Customize Scatter Plot
plt.scatter(df["Study_Hours"], df["Marks"], color="red",
            marker="x", s=100)
plt.title("Customized Scatter Plot")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()

	#•	color="red" → red points
	#•	marker="x" → X-shaped points
	#•	s=100 → bigger size


#Scatter Plot with Two Groups (Colors)
group1 = df[df["Study_Hours"] <= 4]
group2 = df[df["Study_Hours"] > 4]

plt.scatter(group1["Study_Hours"], group1["Marks"], color="blue",
            label="Low Hours")
plt.scatter(group2["Study_Hours"], group2["Marks"], color="green",
            label="High Hours")

plt.title("Scatter Plot with Groups")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.legend()
plt.show()


#Scatter Plot with Trend Line

import numpy as np

x = df["Study_Hours"]
y = df["Marks"]

# Fit a trend line
m, b = np.polyfit(x, y, 1)
plt.scatter(x, y, color="purple")
plt.plot(x, m*x + b, color="black", linestyle="--", label="Trend Line")

plt.title("Scatter Plot with Trend Line")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.legend()
plt.show()


import matplotlib.pyplot as plt

# Sample Data
sizes = [25, 35, 20, 20]
labels = ["Tanya", "Shivani", "Rishi", "Shubham"]

# -----------------------------
# 1. Basic Pie Chart
plt.pie(sizes, labels=labels)
plt.title("Marks Distribution (Basic Pie)")
plt.show()

# -----------------------------
# 2. Pie Chart with Percentages
plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.title("Marks Distribution with Percentages")
plt.show()

# -----------------------------
# 3. Explode (Highlight one slice)
explode = [0.1, 0.2, 0.3, 0.4]  # highlight Tanya's slice
plt.pie(sizes, labels=labels, autopct="%1.1f%%", explode=explode)
plt.title("Highlight Tanya's Slice")
plt.show()

# -----------------------------
# 4. Custom Colors
colors = ["pink", "lightblue", "lightgreen", "orange"]
plt.pie(sizes, labels=labels, autopct="%1.1f%%", colors=colors)
plt.title("Pie Chart with Custom Colors")
plt.show()

# -----------------------------
# 5. Start Angle + Shadow
plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=90, shadow=True)
plt.title("Pie Chart with Start Angle & Shadow")
plt.show()

# -----------------------------
# 6. Donut Chart (Pie with Hole)
plt.pie(sizes, labels=labels, autopct="%1.1f%%")
# draw a white circle in the middle
centre_circle = plt.Circle((0,0),0.70,fc="white")
plt.gca().add_artist(centre_circle)
plt.title("Donut Chart")
plt.show()
```
