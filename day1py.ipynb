```python
# Integer
age = 22
print(type(age))

# Float
height = 165.5
print(type(height))


# String
name = "Tanya"
print(type(name))

# Boolean
is_student = True
print(type(is_student))


# string operation
text = "data science"
print(text.upper())     # Uppercase
print(text.lower())     # Lowercase
print(text.title())     # Title Case
print(text.replace("data", "machine"))  # Replace words
print("science" in text)  # Check if contains word → True/False


# loops

# For loop
for i in range(1, 6):
    print(f"Number: {i}")

# While loop
count = 0
while count < 3:
    print("Hello")
    count += 1

# dictionary
student = {
    "name": "Tanya",
    "age": 22,
    "course": "Data Science"
}

print(student["name"])      # Access value
student["age"] = 23         # Update value
student["city"] = "Raipur"  # Add new key-value
print(student)


# list add insert delete sort


fruits = ["apple", "banana", "cherry"]

# Add
fruits.append("mango")
print(fruits)

# Insert at position
fruits.insert(1, "orange")
print(fruits)

# Remove by value
fruits.remove("banana")
print(fruits)

# Remove last item
fruits.pop()
print(fruits)

# Sort
fruits.sort()
print(fruits)



#list comprehension

# Traditional way
squares = []
for i in range(1, 6):
    squares.append(i * i)
print(squares)

# List comprehension
squares = [i * i for i in range(2, 9)]
print(squares)



#learn if elif else

marks = 85

if marks >= 90:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")
else:
    print("Grade: C")



# learn   Functions with Parameters & Defaults

def greet(name, msg="Welcome to Data Science"):
    print(f"Hello {name}, {msg}")

greet("Tanya")
greet("Ravi", "Good luck with Python!")



# learn importing modules

import math

print(math.sqrt(25))    # Square root
print(math.pi)          # π value
print(math.factorial(5))# Factorial






```
