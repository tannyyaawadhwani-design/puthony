```python
#basic of visualization matplotlib
import matplotlib.pyplot as plt
import pandas as pd

# Sample DataFrame
data = {"Name": ["Tanya", "Shivani", "Rishi", "Shubham"],
        "Marks": [99, 88, 77, 66]}
df = pd.DataFrame(data)


#Line Plot
plt.plot(df["Name"], df["Marks"])
plt.title("Marks of Students")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()


#Bar Plot
plt.bar(df["Name"], df["Marks"])
plt.title("Marks of Students (Bar Chart)")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()


#Scatter Plot
plt.scatter(df["Name"], df["Marks"])
plt.title("Marks Scatter Plot")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()


#Pie Chart
plt.pie(df["Marks"], labels=df["Name"], autopct="%1.1f%%")
plt.title("Marks Distribution")
plt.show()


#MINI CHALLENGE
#Bar Plot
plt.bar(df["Name"], df["Marks"])
plt.title("STUDENT MARKS COMPARISON")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()


#CUSTOMIZE BAR PLOT
#CHANGE BAR COLOR
plt.bar(df["Name"], df["Marks"], color="pink")
plt.title("STUDENT MARKS COMPARISON")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()


#ROTATE X-AXIS,Y-AXIX LABELS
plt.bar(df["Name"],df["Marks"],color="black")
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.title("STUDENT MARKS COMPARISON")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()


#ADD GRIDLINES
plt.bar(df["Name"], df["Marks"], color="green")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.title("Student Marks Comparison")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()

#ADD LABELS ON BARS
bars = plt.bar(df["Name"], df["Marks"], color="purple")
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2,
             yval + 1, yval, ha="center")

plt.title("Student Marks Comparison")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()

# Create bar plot
bars = plt.bar(df["Name"], df["Marks"], color="teal")

# Add labels automatically
plt.bar_label(bars, label_type="edge", padding=2)

# Titles
plt.title("Student Marks Comparison")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()

#Bigger & Bold Labels

bars = plt.bar(df["Name"], df["Marks"], color="orange")

# Customize font size & weight
plt.bar_label(bars, label_type="edge", padding=3,
              fontsize=12, fontweight="bold")

plt.title("Student Marks Comparison")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()

#Add % Instead of Exact Numbers
bars = plt.bar(df["Name"], df["Marks"], color="green")

# Convert labels to percentage of total marks
total = df["Marks"].sum()
percentages = [f"{(mark/total)*100:.1f}%" for mark in df["Marks"]]

plt.bar_label(bars, labels=percentages, label_type="edge",
              padding=3, fontsize=11)

plt.title("Marks as Percentage of Total")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()




```
