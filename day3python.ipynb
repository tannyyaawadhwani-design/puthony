```python
import pandas as pd

# A Series is like one column
fruits = pd.Series(["Apple", "Banana", "Mango", "Grapes"])
print(fruits)
data = {
    "Name": ["Tanya", "Rohit", "Simran", "Aman"],
    "Age": [22, 25, 23, 24],
    "City": ["Raipur", "Delhi", "Mumbai", "Chennai"]
}

df = pd.DataFrame(data)
print(df)
print(df["Name"])     # Access column
print(df.iloc[0])     # Access first row (by index)
print(df.loc[2])      # Access row with index=2
print(df.head(2))        # First 2 rows
print(df.describe())     # Summary of numbers
print(df[df["Age"] > 23])  # Filter condition


#mini exercise
import pandas as pd
students= { "Name": ["Tanya","Shivani","Rishi","Shubham","SRST"],
           "Marks": [99,88,77,66,55], "Subject": ["Python","Php", "Javascript",
                                                  "Java","sql"]}

student_df=pd.DataFrame(students)
print(student_df)
print(student_df["Marks"])
print(student_df.loc[1])
print(student_df[student_df["Marks"]>80])
print("Average Marks:", student_df["Marks"].mean())
print("Max Marks:", student_df["Marks"].max())
print("Min Marks:", student_df["Marks"].min())

#adding column
student_df["Grade"] = ["A","B","C","D","E"]
print(student_df)

#sorting by marks
print(student_df.sort_values("Marks", ascending=False))

#Updating data
student_df.loc[student_df["Name"]=="SRST","Marks"]= 11
print(student_df)

#grouping data
student_df["Department"]=["DS","WB","DS","IT","WB"]
#What is grouping Data average marks per department
print(student_df.groupby("Department")["Marks"].mean())

#descriptive statistics
print(student_df.describe())



#mini exercise
#Adding a new column with condition
student_df["Pass/Fail"]= ["Pass" if m >= 60 else "Fail"
for m in student_df["Marks"]]
print(student_df)

#sorting the data frame by name alphabetically
print(student_df.sort_values("Name", ascending= True ))

#find the student with highest mark
print(student_df[student_df["Marks"] == student_df["Marks"].max()])

#grouping and counting by grade
print(student_df.groupby("Grade")["Name"].count())

#Filtering and multiple condition
print(student_df[(student_df["Marks"] > 70 )
         &  (student_df["Subject"] == "Python")])

#select only name and marks column which condition
#print(student_df[student_df[(student_df["Name"]) & (student_df["Marks"])]
                 #== student_df[student_df["Subject"]== ("Python") | ("Java")]])
print(student_df.loc[student_df["Subject"].isin(["Python", "Java"]),
 ["Name", "Marks"]])


#grouping with condition
print(student_df.groupby("Subject")["Marks"].mean())

#Top two student with highest marks and only their names and marks column
print(student_df.nlargest(2, "Marks")[["Name", "Marks"]])

#thoda tough
print(student_df[student_df["Pass/Fail"].isin(["Pass"])].groupby("Subject")["Marks"].mean())
```
